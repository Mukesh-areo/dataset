```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from bs4 import BeautifulSoup

# Sample list of text content
text_list = ["Downloaded file list", "This application demonstrates the usage of","The application downloads files from"]


with open("D:\Data tracks projects\page_1 .html",'r') as f:
    html_content=f.read()

# Preprocess the text for vectorization
vectorizer = TfidfVectorizer()
text_vectors = vectorizer.fit_transform(text_list)

# Parse HTML content
soup = BeautifulSoup(html_content, 'html.parser')
text_blocks = [tag.get_text() for tag in soup.find_all()]

# Iterate over text blocks
for text_block in text_blocks:
    # Calculate similarity with each text in the list
    text_block_vector = vectorizer.transform([text_block])
    similarities = cosine_similarity(text_vectors, text_block_vector).flatten()

    # Highlight if similarity is above 90%
    for i, similarity in enumerate(similarities):
        if similarity > 0.9:
            for tag in soup.find_all(string=text_block):
                tag.wrap(soup.new_tag('mark', style='background-color: yellow;'))

# Print the modified HTML
print(soup.prettify())
```

```python
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample tagged HTML content from the source file
source_html_content = """
<!-- Sample tagged HTML content -->
<ix:nonNumeric name="eg:DescriptionOfPolicy" context="c1" id="f1" escape="false" continuedAt="cont1">
<div>
    1 - Downloaded file lists<br>
    2 - Start download - small file<br>
    3 - Start download - large file<br>
    Pause - Pause download<br>
    Play - Resume download<br>
    Stop - Cancel download<br>
    0 - Clear logs
</div>
</ix:nonnumeric>
"""

# Sample HTML content from the target file
target_html_content = """
<!-- Sample HTML content -->
<div>
    1 - Downloaded file lists<br>
    2 - Start download and clear- small file<br>
    3 - Start download the content- large file<br>
    Pause - Pause download<br>
    Play - Resume download<br>
    Stop - Cancel download<br>
    0 - Clear logs
</div>
"""

# Parse the source and target HTML content
source_soup = BeautifulSoup(source_html_content, 'html.parser')
target_soup = BeautifulSoup(target_html_content, 'html.parser')

# Extract text from the source and target HTML
source_text = source_soup.get_text()
target_text = target_soup.get_text()

# Calculate similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([source_text, target_text])
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix)[0]

# If similarity is above 90%, tag the text in the target HTML
if cosine_sim[1] > 0.9:
    # Find the target tag in the modified target HTML
    target_tag = target_soup.find('div')
    # Replace the content of the target tag with the source tag content
    target_tag.string = str(source_soup.find('ix:nonnumeric').div)

# Output the modified HTML content
print(target_soup.prettify())
```
